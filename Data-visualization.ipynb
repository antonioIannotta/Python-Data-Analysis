```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1)
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
ax4 = fig.add_subplot(2, 2, 4)
```

```python
fig = plt.figure()
data = np.arange(0.0, 20.0, 0.1)
plt.plot(data, np.sin(data), 'r--')
plt.xlabel('Values')
plt.grid(color='b')
plt.ylabel('Sin Function')
plt.title('Simple example')
plt.xticks([0, 5, 10, 15, 20])
plt.show()
```

```python
general_data = np.arange(0.0, 20.0, 0.1)
plt.plot(general_data, np.sin(general_data), 'r--', label='Sin function')
plt.plot(general_data, np.cos(general_data), 'b--', label='Cos function')
plt.xlabel('Values')
plt.ylabel('Sin function')
plt.title('Simple example')
plt.xticks([0, 5, 10, 15, 20])
plt.show()
```

```python
data = np.arange(0.0, 20.0, 0.5)
plt.plot(data, np.power(data, 2), 'ro-', linewidth=0.1, label='Sqr function')
plt.plot(data, np.log(data), 'bo-', linewidth=0.1, label='Log function')
plt.xlabel('X')
plt.ylabel('F(X)')
plt.show()
```

```python
data0 = np.random.normal(0, 0.5, size=(2, 100))
data1 = np.random.normal(1.0, 0.1, size=(2, 100))
plt.scatter(data0[0], data0[1], c='r', marker='o')
plt.scatter(data1[0], data1[1], c='b', marker='*')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

```python
data = np.array([71.4, 3.1, 0.4, 0.3, 24.8])
label = ['Christianity', 'Islam', 'Buddhism', 'Hinduism', 'Other']
plt.pie(data)
plt.legend(label)
plt.title('Religion in Italy')
plt.show()
```
