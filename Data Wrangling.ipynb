```python
data = pd.DataFrame({'food': ['bacon', 'pulled pork', 'bacon', 'Pastrami'],
                     'ounces': [4, 3, 12, 6]})
data
```

```python
meat_to_animal = {
    'bacon': 'pig',
    'pulled_pork': 'pig',
    'pastrami': 'cow'
}

data['animal'] = data['food'].map(str.lower).map(meat_to_animal)
data
```
